```python
# Frequency Distribution Table
frequency_table = pd.DataFrame(df['ph'].value_counts())
print("Frequency Distribution Table:")
print(frequency_table)
```

```python
# Bar Chart
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(df['ph'], df['Hardness'])
plt.xlabel('pH')
plt.ylabel('Hardness')
plt.title('Hardness Distribution')
plt.show()
```

```python
# Bar Chart
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.bar(df['ph'], df['Conductivity'])
plt.xlabel('pH')
plt.ylabel('Conductivity')
plt.title('Conductivity Distribution')
plt.show()
```

```python
potability_frequency = df['Potability'].value_counts()
print("Potability Frequency Distribution:")
print(potability_frequency)

# Pie Chart for Potability
plt.figure(figsize=(6,6))
plt.pie(potability_frequency, labels=['Non-Potable','Potable'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Potability Distribution (Pie Chart)')
plt.show()
```

```python
# Frequency Polygon
plt.figure(figsize=(8, 5))
plt.hist(df['ph'].dropna(), bins=20, edgecolor='black')
plt.xlabel('pH Values')
plt.ylabel('Frequency')
plt.title('pH Values Distribution (Frequency Polygon)')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(df['ph'], df['Hardness'], alpha=0.5, color='blue')
plt.title('Scatter Plot of pH vs. Hardness')
plt.xlabel('pH')
plt.ylabel('Hardness')
plt.grid(True)
plt.show()
```
